```python
data1 = pd.DataFrame({
    'key1':[1,2,3,4,5],
    'key2':[5,8,4,7,4],
    'key3':[8,5,6,8,9]
})
```

```python
data1
```

```python
data2 = pd.DataFrame({
    'key1':[1,2,46,4,5],
    'key5':[56,8,4,7,4],
    'key6':[1,5,7,8,6]
})
```

```python
data2
```

```python
pd.merge(data1,data2) # by default how= inner(jo dono same hai )  

# colums sane hoga uske same row same same data hai to data dega✅ . row diffent  data nahi dega❌. row same hai us row sab data dega key 2,3,5,6 ka bhi data dega
```

```python
#NOTICE
#left = jo phele diya ho o left hoga(ex-data1)◀
#right = jo bad me diya ho right hoga(ex- data2)▶
#ader left ya right jo bhi hoga uska sab sane data laye 
# left same  hoga ager  right cillums (same) melata hai left se hoga uske same row same same data hai to data dega✅ . row diffent  data nahi dega❌. left row same hai right row sab data same to data dega us row ka . same nahi data nahi to nan dega 
#same right ke sath bhi 
```

```python
pd.merge(data1,data2,how='left') 
```

```python
pd.merge(data1,data2,how='right')
```

```python
pd.merge(data1,data2,how='outer') # dono se data laye ga smae hai data dena same nahi hoga ak ak data(jis ka data hoga) dega
```

```python
pd.merge(data1,data2,how='outer',on="key1") # on kiya to kese marge karva chate 
```

```python
pd.merge(data1,data2,how='cross') 

#compare kar raha data1 ak row le raha than data2 ka sab row compare kae raha bhir data1 second row phir ta2 ka sab row compare kae chale jata hai 
```

```python
pd.merge(data2,data1,how='cross') # jo phele usk likha hoga compare hoaga
#phele hoga  uska ak row lega bad wala me  us row ke sath sab row compare hoga

#compare kar raha data2 ak row le raha than data1 ka sab row compare kae raha bhir ) 
```
