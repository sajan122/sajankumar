```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
fig = go.Figure(data=[go.Histogram(x = tips.total_bill)])
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x =tips.total_bill , y =tips.tip,mode='lines'))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x =tips.total_bill , y =tips.tip,mode='markers'))
fig.show()
```

```python
tips.head()
```

```python
import pandas as pd
tips['size'].value_counts()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x =tips.total_bill , y =tips.tip,mode='markers',marker=dict(size = 10 * tips['size'],color=tips['size'], showscale = True) )) # 1 loga group than size chota 2 loga groph use thoda bada ______
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x =tips.total_bill , y =tips.tip, z=tips['size'],mode='markers' )) 
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x =tips['total_bill'] , y =tips['tip'], z=tips['size'],mode='markers' )) #3D points tabhi clearly dikhte hain jab tum marker dict use karo
fig.show()
```
